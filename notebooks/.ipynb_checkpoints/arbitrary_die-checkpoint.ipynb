# Investigations on Rolling Dice
A lot of probabilistic insight can be gained from the study of a die roll. Typically, we concern ourselves
with a 6-sided die and ask questions along the lines of "what is the probability we observe 3 given a roll
of the die?"

Let's investigate some mathematical and computational aspects of rolling die/dice. The first, trivial, question I want to investigate is determining the probability of observing some face on the die given
a single roll. Let the random variable $D$ be the roll of a fair die. Then, we can denote the probability of $D = d$ as follows:
$$
Pr(D=d) = \frac{1}{6}
$$
and we can simulate this in a fairly straightforward manner. Also note the expected value is easily derived as follows:
$$
\mathbb{E}[D] = \sum_{d=1}^{6} d * Pr(D=d) = \frac{15}{6}
$$

```julia
using Plots
```

```julia
MAX_ROLLS = 1000
EV = 3.5

# Simulate die roll for MAX_ROLLS
die_roll() = rand(1:6)
rolls = [die_roll() for i in 1:MAX_ROLLS]
cum_avg = []

# Compute the cumulative average
for i in 1:MAX_ROLLS
    push!(cum_avg, sum(rolls[1:i]) / i)
end
```

```julia
plot(cum_avg, ylims=(3., 4.), label="Cumulative Average", xlabel="# of Simulations")
hline!([EV], label=nothing)
```
